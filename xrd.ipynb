```python
from mp_api.client import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
import matplotlib.pyplot as plt
```

```python
mp_key = "PgRtTbiMJ5N3guKB8xgaTcXoz0bUQyKD"
mp_id = "mp-15788"
```

```python
with MPRester(mp_key) as m:
    prim_struc = m.get_structure_by_material_id(mp_id)
```

```python
print("Voici le diffractogramme pour un rayonnement X incident Cu Kα de longueur d’onde λ = 1.54060 Å :")
xrd = XRDCalculator(1.54060)
dif = xrd.get_pattern(prim_struc)
fig = xrd.get_plot(prim_struc, annotate_peaks=False, two_theta_range=[0, 180])
plt.show()

print("Et voilà les 3 premiers pics du diffractogramme correspondant :")
fig2 = xrd.get_plot(prim_struc,annotate_peaks=False,two_theta_range=[10,29])
plt.show()
```
