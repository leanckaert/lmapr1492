# 1) Rotation

```python
image_1=img.imread('LiYS2-1.png') #image provenant de JSmol
fig, axes = plt.subplots(figsize=(8,8))
axes.get_xaxis().set_visible(False)
axes.get_yaxis().set_visible(False)
axes.imshow(image_1)
plt.show()
```

# 2) Réflexion

```python
image_2=img.imread('LiYS2-2.png') #image provenant de JSmol
fig, axes = plt.subplots(figsize=(8,8))
axes.get_xaxis().set_visible(False)
axes.get_yaxis().set_visible(False)
axes.imshow(image_2)
plt.show()
```

```python
print("Atome : Y")
print("Opération : Réflexion -> -y, -x, z")
print("Description : Réflexion à travers un plan miroir vertical. C'est une opération qui inverse la chiralité.")
print("Position initiale : [0, 0, 1]")
print("Position finale : [0, 0, 1]")
print("Etant donné que les coordonnées x et y de l'atome sont nulles, les positions initiale et finale sont identiques.")
```
